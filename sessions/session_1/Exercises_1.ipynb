### Class

Python is an object-oriented language, which allows us to structure the code using classes and objects for better organization and reusability.

For example, a class could be Animal and an object of that class would be dog. In a class, you have the `__init__` method, that allows you to initialize an object given a set of parameters. With those parameters, you initialize the internal *attributes* of the class, e.g., `self.something`. You can also define internal methods for the class, which are functions you can call if you have declared an object of that class.

```python
class Animal:
  def __init__(self, name: str, legs: int, sound: str):
    # Initialization of internal parameters
    self.name = name
    self.legs = legs
    self.sound = sound

  def make_sound(self):
    return self.sound + "!"
```

To access the attributes or methods, you can just call them after using a dot on the declared object.

```python
dog = Animal("dog", 4, "woof")
print(dog.name)
print(dog.legs)
print(dog.make_sound())
```

### Tensors: The building blocks of PyTorch





A tensor is a multi-dimensional array that can be used to store data. It is the fundamental building block of PyTorch, and it is used to represent the input data, the parameters of the model, and the output of the model. Tensors can be of different types, such as `torch.FloatTensor`, `torch.IntTensor`, etc., depending on the type of data they store. We can perform various operations on tensors, such as addition, multiplication, etc. They allow us to perform operations parallelly on the GPU, on all the elements of the tensor at the same time, or between tensors if they have the same shape. This is what makes PyTorch so powerful for Deep Learning, as it allows us to perform operations on large amounts of data efficiently.

```python
import torch

# Create a tensor of size (2, 3) filled with random values
tensor_1 = torch.randn(2, 3)
print(tensor_1)
```

```python
# Create a tensor of size (2, 3) with the values [[1, 2, 3], [4, 5, 6]]
tensor_2 = torch.tensor([[1, 2, 3], [4, 5, 6]])
print(tensor_2)
```

```python
# Sum of two tensors
print(tensor_1 + tensor_2)
```

```python
# Multiplication of two tensors
print(tensor_1 * tensor_2)
```

### Defining the Architecture: The `nn.Module` class

In Python, we have can have inheritance relationships between classes. A class that inherits from another class is called a child class, and the class that is inherited from is called a parent class. The child class can use all the methods and attributes of the parent class, and it can also have its own methods and attributes.

In PyTorch, we define the architecture of a model by creating a child class of `nn.Module`, which is the parent class that contains all the necessary methods and attributes to work with PyTorch. In this child class, we have to define the `__init__` method, where we initialize the layers of the model, and the `forward` method, where we define how the input data flows through the layers to produce the output.

```python
import torch
import torch.nn as nn

class MyModel(nn.Module):
  def __init__(self):
    super(MyModel, self).__init__()
    self.layer1 = nn.Linear(10, 5)
    self.layer2 = nn.Linear(5, 1)

  def forward(self, x):
    x = self.layer1(x)
    x = torch.relu(x)
    x = self.layer2(x)
    return x

model = MyModel()
```

```python
vector = torch.randn(10)  # Random input vector of size 10
output = model(vector)
print(output)
```

Using PyTorch, we can separate the layers from the activation functions, which gives us more flexibility to define the architecture of the model. In the example above, we have defined just two layers (`layer1` receives an input of size 10 and produces an output of size 5, and `layer2` receives an input of size 5 and produces an output of size 1), and we have used the ReLU activation function between them. We can easily change the architecture by adding more layers or using different activation functions.

### The Loss Function

The loss function is a measure of how well the model is performing. It takes the output of the model and the true labels, and it returns a single value that represents the error of the model.

PyTorch has a lot of built-in loss functions that we can use, such as `nn.MSELoss` for regression problems and `nn.CrossEntropyLoss` for classification problems. We can also define our own custom loss functions if we need to. If that is the case, they also inherit from `nn.Module` and we have to define the `forward` method, where we calculate the loss given the output of the model and the true labels.

```python
import torch.nn as nn

mse_loss = nn.MSELoss()
input = torch.tensor([0.0, 0.0, 0.0, 0.0])
output = torch.tensor([1.0, 1.0, 1.0, 1.0])
loss = mse_loss(input, output)
print(loss)
```

```python
class CustomLoss(nn.Module):
  def __init__(self):
    super(CustomLoss, self).__init__()

  def forward(self, input, output):
    return torch.sum((input - output) ** 2) / input.size(0)  # This is just an example, another MSE loss

custom_loss = CustomLoss()
loss = custom_loss(input, output)
print(loss)
```

### The Training Loop

As we have seen in the lecture, the gradient descent can be programmed in a loop, since we have to update the weights of the model several times until we reach a good performance. The training loop consists of the following steps:
1. Forward pass: we pass the input data through the model to get the output.
2. Calculate the loss: we calculate the loss using the output of the model and the true labels.
3. Backward pass: we calculate the gradients of the loss with respect to the parameters of the model using backpropagation.
4. Update the weights: we update the weights of the model using the gradients calculated in the backward pass and an optimization algorithm, such as Stochastic Gradient Descent (SGD) or Adam (we will see these in the next session).

PyTorch does all of these steps for us, so we just have to write the code to call the necessary methods and functions.

```python
import torch

model = MyModel()
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)

for epoch in range(100):  # Loop over the dataset multiple times
    # Forward pass
    output = model(vector)

    # Calculate the loss
    loss = criterion(output, output)  # This is just an example, we should use the true labels instead of the output

    # Backward pass
    optimizer.zero_grad()  # Zero the gradients before backward pass
    loss.backward()  # Compute the gradients

    # Update the weights
    optimizer.step()  # Update the weights using the gradients
```
